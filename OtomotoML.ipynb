```python
import os
import pandas as pd
import numpy as np

current_directory = os.path.join(os.getcwd(), "data")

dfs = []

for file in os.listdir(current_directory):
    if file.endswith('.csv'):
        file_path = os.path.join(current_directory, file)
        df = pd.read_csv(file_path)
        dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)

merged_df
```

```python
df = merged_df.drop(columns=['Offer Title', 'Car Link', 'Version', 'Model', 'City Fuel Consumption'])
df.info()
```

```python
def unique_values_for_columns(df):
    unique_values_dict = {}
    for column in df.columns:
        unique_values_dict[column] = df[column].unique()
    for column, values in unique_values_dict.items():
        print(f"Unique values for column '{column}': {values}")

unique_values_for_columns(df)
```

```python
def is_number(x):
    try:
        float(x)
        return True
    except ValueError:
        return False

non_numeric_prices = df[~df['Price'].apply(is_number)]['Price']
print(non_numeric_prices.unique())
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df = df.dropna(subset=['Price'])
print(df.info())
```

```python
df['Mileage'] = df['Mileage'].str.replace(' ', '')

df['Mileage'] = df['Mileage'].str.replace('km', '')

df['Mileage'] = df['Mileage'].astype(float)
print(df.info())
```

```python
df['Engine Capacity'] = df['Engine Capacity'].str.replace(' ', '')

df['Engine Capacity'] = df['Engine Capacity'].str.replace('cm3', '')

df['Engine Capacity'] = df['Engine Capacity'].astype(float)

print(df.info())
```

```python
df['Power'] = df['Power'].str.replace(' ', '')

df['Power'] = df['Power'].str.replace('KM', '')

df['Power'] = df['Power'].astype(float)
print(df.info())
```

```python
df['Registered in Poland'].fillna('Nie', inplace=True)
df['Accident Free'].fillna('Nie', inplace=True)
df['Tuning'].fillna('Nie', inplace=True)
df['Registration Number'].fillna('Nie', inplace=True)

print(df.info())
```

```python
unique_values_for_columns(df)
```

```python
def createDummiesObj(df, column):
    if (df[column].isna().sum() > 0):
        df[column].fillna(column + "NAN", inplace = True)
    dummies = pd.get_dummies(df[column], prefix=column, dtype=float)
    df = df.drop(column, axis=1)
    df = pd.concat([df, dummies], axis=1, join='inner')
    return df

def fillNanInts(df, column):
    df[column + "NAN"] = np.where(df[column].isnull(), 0, 1)
    df[column].fillna(df[column].mean(), inplace = True)
    return df

def preprocessData(df):
    for column, dtype in df.dtypes.items():
        if dtype == 'object':
            df = createDummiesObj(df, column)
        if (dtype == 'float64' or dtype == 'int64') and (df[column].isna().sum() > 0):
            df = fillNanInts(df, column)
    return df
```

```python
df = preprocessData(df)
df
```

```python
from sklearn.model_selection import train_test_split
#from sklearn.linear_model import  LinearRegression,Lasso,SGDRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Price']), df['Price'], test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators=128, n_jobs=-1) 

model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

train_score = r2_score(y_train, y_train_pred)
test_score = r2_score(y_test, y_test_pred)

print("Train Score (R^2):", train_score)
print("Test Score (R^2):", test_score)
```
